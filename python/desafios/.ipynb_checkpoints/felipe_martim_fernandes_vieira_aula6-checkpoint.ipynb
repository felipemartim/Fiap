## 1) Utilizando o dataset Pokemon, crie uma função que receba o dataset e o atributo e retorne um histograma que mostre a distribuição do valor dos atributos “attack”, “defense”,”Sp.Atk”, “Sp.Def” e “Speed”. Insira uma linha indicando o valor médio do atributo

```python
# resposta 1
import pandas as pd
import matplotlib.pyplot as plt

poke = pd.read_csv('Pokemon.csv')

def pokemon_histogram(dataset, attr):
    plt.hist(dataset[attr], 30)
    plt.text(79, .0170, r'$\mu='+ str(dataset[attr].mean()) +'$')
    
pokemon_histogram(poke, "Attack")
```

## 2) Usando o dataset Pokemon, crie um scatter plot comparando ataque (eixo x) e defesa (eixo y) de pokemons dos tipos Fire e Water. 

```python
# resposta 2
df_fire = poke[poke['Type 1']=='Fire']
df_water = poke[poke['Type 1']=='Water']

plt.scatter(df_fire['Attack'],  df_fire['Defense'],  c= 'seagreen',      label = 'fire')
plt.scatter(df_water['Attack'], df_water['Defense'], c= 'rebeccapurple', label = 'Water')

plt.xlabel('Defesa')
plt.ylabel('Ataque')
plt.legend()

plt.show()
```

## 3) Usando o dataset Pokemons, crie um gráfico de linha para comparar a quantidade de pokemons de cada tipo ao longo das gerações

```python
# resposta 3
df_grouped = pd.DataFrame({'count' : poke.groupby(['Type 1','Generation']).size()}).reset_index()
plt.figure(figsize=(10,10)) 

for p_type in df_grouped['Type 1'].unique():
  df_type = df_grouped[df_grouped['Type 1']== p_type]
  plt.plot(df_type['Generation'], df_type['count'], label=p_type)
  plt.xlabel('Geração')
  plt.ylabel('Qtde')
  plt.title("Quantidade de tipos de pokémon por geração")
  plt.legend()
```

## 4) Utilizando o dataset airports (var = data.airports()):
    -> Faça um rank bar dos 5 primeiros estados dos EUA com maior número de aeroportos


```python
# resposta 4
import altair as alt
from vega_datasets import data

df_airports = data.airports()
df_airports.head()

alt.Chart(df_airports).transform_aggregate(
    count='count()',
    groupby=['state']
).transform_window(
    rank='rank(count)',
    sort=[alt.SortField('count', order='descending')]
).transform_filter(
    alt.datum.rank < 6
).mark_bar().encode(
    y=alt.Y('state:N',
        sort=alt.EncodingSortField(field='count', op='sum', order='descending')
    ),
    x='count:Q',
)
```

## 5) Utilizando o conjunto de dados íris (var = data.íris()):
    -> Crie dois gráficos scatter, chart1 e chart2. Para o chart1, x=petalLenght, y=petalWidth. Para o chart2, x = sepalLenght, y=sepalWidth.

    -> Componha esses gráficos numa única apresentação. Para isso, faça chart1 | chart2


```python
# resposta 5
df_iris = data.iris()

brush = alt.selection(type='interval')

chart1 = alt.Chart(df_iris).mark_point().encode(
    x='petalLength:Q',
    y='petalWidth:Q',
    color=alt.condition(brush, 'species:O', alt.value('grey'))
).add_selection(brush)

chart2 = alt.Chart(df_iris).mark_point().encode(
    x='sepalLength:Q',
    y='sepalWidth:Q',
    color=alt.condition(brush, 'species:O', alt.value('grey'))
).add_selection(brush)

chart1 | chart2
```

## 6) Usando o dataset Pokemon, crie um rank bar que mostre o rank para Type 1

```python
# resposta 6
alt.Chart(poke).transform_aggregate(
    count='count()',
    groupby=['Type 1']
).transform_window(
    rank='rank(count)',
    sort=[alt.SortField('count', order='descending')]
).mark_bar().encode(
    y=alt.Y('Type 1:N',
        sort=alt.EncodingSortField(field='count', op='sum', order='descending')
    ),
    x='count:Q',
)
```

## 7) Utilizando o dataset Pokemon, crie um gráfico de barras que mostre, para cada geração, a proporção de pokemons Legendários e aqueles que não são legendários. Use stacked bar. 

```python
# resposta 7
barras = alt.Chart(poke).mark_bar().encode(
    x='count()',
    y='Generation',
    color='Legendary'
)

texto = alt.Chart(poke).mark_text(dx=-15, dy=3, color='white').encode(
    x=alt.X('count()):Q', stack='zero'),
    y=alt.Y('Generation:N'),
    detail='Legendary:N',
    text=alt.Text('count():Q', format='.1f')
)

barras + texto
```
